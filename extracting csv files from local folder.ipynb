```python
class FoldersInThePath:#Initialising class creating constructor
    def __init__(self,path):
        self.logging=logit('logger_for_data_extraction')
        self.create_column_type='none'
        self.path=path
        os.chdir(self.path)          
        self.list=os.listdir()           
    def isitdir(self,newpath):#check if it is directory'''
        try:
            if os.path.isdir(newpath)==True:
                self.logging.info('checked  file is directory ')
                return True
            else:
                self.logging.info('checked  file is not directory ')
                return False
        except Exception as e:
            self.logging.info(str(e),'error in checking if it it is directory')
            return (str(e),'error in checking if it it is directory')
    def changepath(self,cpath):
        #change the path to given path
        try:
            os.chdir(cpath)
            self.logging.info('path changed')
            return True
        except Exception as e:
            self.logging.info(str(e),'error in changing path')
            return (str(e),'error in changing path')
    def dir_list(self):
        #list of directories in current folder
        try:
            self.logging.info('directories in current folder listed')
            return os.listdir()
        except Exception as e:
            self.logging.info(str(e),'error in listing the directories')
            return (str(e),'error in listing the directories')
    def joinit(self,parent,new):
        #joining 2 tables
        try:
            joined=pd.concat([parent,new])
            self.logging.info('joined the tables')
            return joined
        except Exception as e:
            self.logging.info(str(e),'error in joining')
            return (str(e),'error in joining')
    def read_csv_file(self,extract_path):
        # reading csv file and creating column
        try:
            data=pd.DataFrame(pd.read_csv(extract_path,header=4))
            self.logging.info('csv file read')
            data['type']=str(self.create_column_type)
            self.logging.info('column created')
            return data
        except Exception as e:
            self.logging.info(str(e),'error in reading csv file')
            return (str(e),'error in reading csv file')
    def get_current_path(self):
        #get current path
        try:
            path=os.getcwd()
            self.logging.info('current path is saved in the variable '+str(path))
            return path
        except Exception as e:
            self.logging.info(str(e),'error in getting current path')
            return (str(e),'error in getting current path')
    def extract_csv_files(self,path):
        #extracting data from csv file
        try:
            parent=pd.DataFrame()
            self.logging.info('empty dataframe is created '+str(parent))
            for files in FoldersInThePath.dir_list(self):
                self.logging.info('iterating through list of current folders extracting from csv files')
                if files.endswith('.csv'):#checked if current folders end with csv
                    self.logging.info('checked if current folders end with csv')
                    extract_path=FoldersInThePath.get_current_path(self)+r'\\' + str(files)
                    self.logging.info('called function '+('get_current_path'))
                    extracted_data=FoldersInThePath.read_csv_file(self,extract_path)
                    self.logging.info('called function '+('read_csv_file'))
                    parent=FoldersInThePath.joinit(self,parent,extracted_data)
                    self.logging.info('called function '+('joinit'))
                    self.logging.info('returning extracted file')
            return parent
        except Exception as e:
            self.logging.info(str(e),'error in extracting csv file')
            return (str(e),'error in extracting csv file')
    def main_program(self):
        try:# create empty dataframe to tore data
            final_data=pd.DataFrame()
            self.logging.info('creating final dataframe'+str(final_data))
            for file1 in self.list:#checking folders within given folder
                self.logging.info('iterating through list of current folders wherein each folder contains csv files')
                newpath=path+file1#creating new path for changing current path
                self.logging.info('new path created'+str(newpath))
                self.create_column_type=file1 #giving filename so as to create column in the data 
                self.logging.info('giving value to new column'+str(self.create_column_type))
                if FoldersInThePath.isitdir(self,newpath) is True:#iterate through dictionaries
                    FoldersInThePath.changepath(self,newpath)  #change path for extracting data from that file
                    self.logging.info('new path created'+('newpath'))
                    parent_data=FoldersInThePath.extract_csv_files(self,newpath)#extract csv data
                    self.logging.info('called function '+('extract_csv_files'))
                    final_data=FoldersInThePath.joinit(self,final_data,parent_data)#joining extracted data to table
                    self.logging.info('final data created by calling function '+('joinit'))
            return final_data   #return final data
        except Exception as e:
            self.logging.info(str(e),'error in main program')
            return (str(e),'error in main program')
           
```
